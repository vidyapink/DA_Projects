```python
# After your age cleaning and filtering
df['age'].hist(bins=20, edgecolor='pink')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.show()
```

```python
### Which age group is having more stress level
```

```python
# Create the bar plot using Seaborn
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='age', y='stress_level', estimator='mean', errorbar=None)

# Add labels and title
plt.xlabel('age')
plt.ylabel('Stress Level')
plt.title('Average Stress Level by Age Group')
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.barplot(data=df,x='gender',y='stress_level',estimator='mean',errorbar=None)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)

plt.xlabel('age')
plt.ylabel('Stress level')
plt.title('Average stress level by Gender')

plt.show()
```

```python
# which platform is used more
plt.figure(figsize=(4,5))
ax=sns.barplot(data=df,x='platform_usage',y='age',estimator='count',errorbar=None)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)
    
plt.show()
```
